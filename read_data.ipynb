## Sensors data

### inbiot

* Data ranges from: **2024-10-11 : 2024-10-21**

```python
df_inbiot = pd.read_excel('data/Inbiot.xlsx')

df_inbiot
```

#### preprocessing

```python
df_inbiot['Date'] = df_inbiot['Date'].astype(str)
df_inbiot['Time'] = df_inbiot['Time'].astype(str)

df_inbiot['datetime'] = pd.to_datetime(df_inbiot['Date'] + ' ' + df_inbiot['Time'], format='%Y-%m-%d %H:%M:%S')


df_inbiot.columns = (
    df_inbiot.columns
    .str.replace(r"\(.*?\)", "", regex=True)  
    .str.replace(r"\s+", "_", regex=True)     
    .str.rstrip("_")                          
    .str.lower()                              
)

df_inbiot["sensor"] = "inbiot"
print(df_inbiot.columns)
df_inbiot.drop(columns=["date","time"],inplace=True)

df_inbiot.set_index("datetime",inplace=True)
```

### Smart sense

```python
df_smart_sense = pd.read_csv('data/smart_sense.csv')

df_smart_sense
```

#### prep

```python
# Standardize column names
df_smart_sense.rename(columns={
    'ts[datetime]': 'start_time',
    'te[datetime]': 'end_time',
    'CO2[ppm]': 'co2',
    'T[°C]': 'temperature',
    'RH[%]': 'humidity',
    'indoor.NOx_index[idx]': 'indoor_nox_index',
    'indoor.VOC_index[idx]': 'indoor_voc_index',
    'PM1[ugm3]': 'pm_1.0',
    'PM2.5[ugm3]': 'pm_2.5',
    'PM10[ugm3]': 'pm_10',
}, inplace=True)


df_smart_sense['start_time'] = pd.to_datetime(df_smart_sense['start_time'],format="%d/%m/%Y %H:%M")
df_smart_sense['end_time'] = pd.to_datetime(df_smart_sense['end_time'],format="%d/%m/%Y %H:%M")

are_equal = df_smart_sense['end_time'].equals(df_smart_sense['end_time'])

assert are_equal == True

df_smart_sense.drop('end_time',axis=1,inplace=True)
```

```python

df_smart_sense.set_index('start_time', inplace=True)
df_smart_sense.rename_axis('datetime', inplace=True)

df_smart_sense["sensor"] = "smart_sense"
df_smart_sense.head()
```
