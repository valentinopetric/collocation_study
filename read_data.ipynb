### inbiot

* Data ranges from: **2024-10-11 : 2024-10-21**

```python
df_inbiot = pd.read_csv('data/Inbiot.csv')

df_inbiot
```

#### preprocessing

```python
df_inbiot['Date'] = df_inbiot['Date'].astype(str)
df_inbiot['Time'] = df_inbiot['Time'].astype(str)

df_inbiot['datetime'] = pd.to_datetime(df_inbiot['Date'] + ' ' + df_inbiot['Time'], format='%Y-%m-%d %H:%M:%S')


df_inbiot.columns = (
    df_inbiot.columns
    .str.replace(r"\(.*?\)", "", regex=True)  
    .str.replace(r"\s+", "_", regex=True)     
    .str.rstrip("_")                          
    .str.lower()                              
)

df_inbiot["sensor"] = "inbiot"
print(df_inbiot.columns)
df_inbiot.drop(columns=["date","time"],inplace=True)

df_inbiot.set_index("datetime",inplace=True)
```

### thinnect

```python
df_full_thinnect = pd.DataFrame()
base_dir = 'data/thinnect'

for pollutant in os.listdir(base_dir):
    df_pol = pd.DataFrame()
    
    for file in os.listdir(os.path.join(base_dir,pollutant)):
        df_mini = pd.read_csv(os.path.join(base_dir,pollutant,file),index_col=0,parse_dates=True)
        df_pol=pd.concat([df_pol,df_mini],axis=0)
   
    df_full_thinnect = pd.concat([df_full_thinnect,df_pol.resample('15min').mean()],axis=1)
    

df_full_thinnect
        
```
